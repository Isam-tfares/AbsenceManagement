<h3>face rec from webcam faster</h3>

```python
known_face_encodings = []
known_face_names = [
#     "Barack Obama",
#     "Joe Biden",
    "Lionel Messi",
    "Isam Tfares"
]

# Load images and encode known faces
def load_known_faces():
#     obama_image = face_recognition.load_image_file("./dataset/obama.jpeg")
#     obama_face_encoding = face_recognition.face_encodings(obama_image)[0]
#     known_face_encodings.append(obama_face_encoding)
#     obama_image = face_recognition.load_image_file("./dataset/biden.jpeg")
#     obama_face_encoding = face_recognition.face_encodings(obama_image)[0]
#     known_face_encodings.append(obama_face_encoding)
    obama_image = face_recognition.load_image_file("./dataset/messi.jpeg")
    obama_face_encoding = face_recognition.face_encodings(obama_image)[0]
    known_face_encodings.append(obama_face_encoding)
    obama_image = face_recognition.load_image_file("./dataset/isam.jpeg")
    obama_face_encoding = face_recognition.face_encodings(obama_image)[0]
    known_face_encodings.append(obama_face_encoding)


load_known_faces()

video_capture = cv2.VideoCapture(0)
while True:
    # Grab a single frame of video
    ret, frame = video_capture.read()

    # Find all face locations in the frame
    face_locations = face_recognition.face_locations(frame)

    # Encode face encodings for all faces found
    face_encodings = face_recognition.face_encodings(frame, face_locations)

    # Initialize an empty list to store names of recognized faces
    face_names = []

    # Check if each face matches any known face
    for face_encoding in face_encodings:
        # Compare the face encoding with known face encodings
        matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
        name = "Unknown"

        # If a match is found, use the name of the known face
        if True in matches:
            first_match_index = matches.index(True)
            name = known_face_names[first_match_index]

        # Append the name to the list of recognized face names
        face_names.append(name)

    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        # Draw a rectangle around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with the name below the face
        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)

    # Display the resulting image
    cv2.imshow('Video', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam
video_capture.release()
cv2.destroyAllWindows()
```
